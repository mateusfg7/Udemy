```
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]
titulo = "Grafico de Barras"
eixo_x = "Eixo X"
eixo_y = "Eixo Y"

# Legendas
plt.title(titulo)
plt.xlabel(eixo_x)
plt.ylabel(eixo_y)

plt.bar(x, y) # Gráfico de Barras
plt.show()
```

```
# Comparar Barras
x1 = [1, 3, 5, 7, 9]
y1 = [2, 3, 7, 1, 0]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 1, 3, 7, 4]

titulo = "Grafico de Barras 2"
eixo_x = "Eixo X"
eixo_y = "Eixo Y"

# Legendas
plt.title(titulo)
plt.xlabel(eixo_x)
plt.ylabel(eixo_y)

plt.bar(x1, y1, label = "Grupo 1")
plt.bar(x2, y2, label = "Grupo 2")
plt.legend() # Plotar as legendas ("Grupo 1" e "Grupo 2")
plt.show()
```

```
# Crescimento da população Brasileira 1980-2016

import pandas as pd

uri = "https://raw.githubusercontent.com/mateusfg7/Udemy/master/populacao-brasileira.csv"
dados = pd.read_csv(uri)

x = []
y = []

for i in range(len(dados)):
  x.append(dados["ano"][i])
  y.append(dados["population"][i])

plt.title("Crescimento da população Brasileira 1980-2016")
plt.xlabel("Ano")
plt.ylabel("população x100.000.000")
plt.bar(x,y, color="#e4e4e4")
plt.plot(x,y, color="k", linestyle="--")
plt.show()
```

```
# Boxplot - Diagrama com caixas

import random

vetor = []

for i in range(10):
  numero_aleatorio = random.randint(0,5)
  vetor.append(numero_aleatorio)

plt.boxplot(vetor) # boxplot
plt.title("Boxplot")
plt.show()
```
